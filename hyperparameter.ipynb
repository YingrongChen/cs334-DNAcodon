```python
import pandas as pd
from sklearn.metrics import accuracy_score, classification_report
datadir = './relabeled_data/data'

X_train = pd.read_csv(f'{datadir}X_train.csv')
y_train = pd.read_csv(f'{datadir}y_train.csv')
X_test = pd.read_csv(f'{datadir}X_test.csv')
y_test = pd.read_csv(f'{datadir}y_test.csv')


X_train_norm = X_train.drop(['Ncodons'], axis = 1)
X_test_norm = X_test.drop(['Ncodons'], axis = 1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

knn = KNeighborsClassifier()
params = {
    'n_neighbors': [1,3,5,7,9,11,13],
    'weights': ['uniform', 'distance'],
    'p': [1,2]
}

clf = GridSearchCV(
    estimator=knn,
    param_grid=params,
    cv=5,
    n_jobs=5,
    verbose=1
)
clf.fit(X_train_norm, y_train)
print(clf.best_params_)
```

```python
from sklearn.ensemble import RandomForestClassifier
import numpy as np

random_grid = {
 'max_depth': [10, 50, 100, 200],
 'min_samples_leaf': [1, 2, 4],
 'n_estimators': [400, 800, 1200, 1600, 1800]}

rf = RandomForestClassifier()
clf = GridSearchCV(
    estimator=rf,
    param_grid=random_grid,
    cv=5,
    n_jobs=2,
)
clf.fit(X_train_norm, y_train)
print(clf.best_params_)
```

```python
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(max_iter=100)
# defining parameter range 
MLP_grid = {
    'hidden_layer_sizes': [(64,600,600,300,150), (50,100,50), (100,)]
}
  
clf = GridSearchCV(mlp, MLP_grid,     cv=5,
    n_jobs=5,
    verbose=1)
  
clf.fit(X_train_norm, y_train)
print(clf.best_params_)
```
